```python
import matplotlib.pyplot as plt

import pandas as pd
import numpy as np
```

2. Create a Data Frame

```python
data ={
    'Country': ['India', 'Japan', 'Nepal', 'Bihar'],
    'Population': [8.5, 4.0, 2.7, 5.5],
    'GDP': [98, 34, 22, 100],
    'Area': [50, 10, 20, 100]
}

df = pd.DataFrame(data)
print(df)
```

3. Line Graph with Labels

```python
x = df['Country']
y = df['Population']

plt.plot(x , y)

plt.xlabel("Countries")
plt.ylabel("Population (in Millions)")
plt.title("Country vs. Population")
plt.grid(True)

plt.show()
```

4. Scatter Plot with Labels

```python
x = df['Country']
y = df['Population']
z = df['GDP']

plt.scatter(x , y)

plt.scatter(x , z)

plt.show()
```
